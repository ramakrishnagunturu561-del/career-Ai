```python
def clean_text(text):

    text = str(text).lower()

    text = " ".join(
        text.split()
    )

    return text


career_df["clean_skills"] = (
    career_df["skills"]
    .apply(clean_text)
)

career_df[
    ["role", "clean_skills"]
].head()
```

```python
model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)

print("NLP model loaded successfully!")
```

```python
career_embeddings = model.encode(
    career_df["clean_skills"].tolist(),
    normalize_embeddings=True,
    show_progress_bar=True
)

print(
    "Embeddings shape:",
    career_embeddings.shape
)
```

```python
def predict_career(
    candidate_skills,
    top_n=5
):

    candidate_text = clean_text(
        candidate_skills
    )

    candidate_embedding = model.encode(
        [candidate_text],
        normalize_embeddings=True
    )

    similarities = cosine_similarity(
        candidate_embedding,
        career_embeddings
    )[0]

    results = career_df[
        ["role", "skills"]
    ].copy()

    results["match_score"] = (
        similarities * 100
    )

    results = results.sort_values(
        "match_score",
        ascending=False
    )

    return results.head(top_n)
```

```python
def get_missing_skills(
    candidate_skills,
    target_role
):

    role_data = career_df[
        career_df["role"] == target_role
    ]

    if role_data.empty:
        return []

    required_text = (
        role_data.iloc[0]["skills"]
        .replace("\n", " ")
    )

    candidate_lower = (
        candidate_skills.lower()
    )

    required_skills = [
        skill.strip()
        for skill in required_text.split()
        if len(skill.strip()) > 2
    ]

    missing = []

    for skill in required_skills:

        if skill.lower() not in candidate_lower:

            if skill not in missing:
                missing.append(skill)

    return missing[:10]
```

```python
import random
import pandas as pd

random.seed(42)

career_skill_map = {

    "AI ML Engineer": [
        "Python", "Machine Learning", "Deep Learning",
        "Scikit-learn", "TensorFlow", "PyTorch",
        "Pandas", "NumPy", "Artificial Intelligence",
        "NLP", "Computer Vision", "Model Deployment"
    ],

    "Data Scientist": [
        "Python", "Machine Learning", "Statistics",
        "Probability", "Pandas", "NumPy",
        "SQL", "Scikit-learn", "Data Visualization",
        "Feature Engineering", "Predictive Modeling"
    ],

    "Data Analyst": [
        "SQL", "Excel", "Python", "Pandas",
        "Power BI", "Tableau", "Statistics",
        "Data Analysis", "Data Visualization",
        "Business Intelligence"
    ],

    "Generative AI Engineer": [
        "Python", "Generative AI", "LLM",
        "Prompt Engineering", "RAG", "LangChain",
        "Vector Databases", "Embeddings",
        "Transformers", "Gemini", "OpenAI",
        "Ollama", "FastAPI"
    ],

    "NLP Engineer": [
        "Python", "NLP", "Transformers",
        "BERT", "Hugging Face", "Machine Learning",
        "Deep Learning", "Text Classification",
        "Embeddings", "PyTorch"
    ],

    "Computer Vision Engineer": [
        "Python", "Computer Vision", "OpenCV",
        "YOLO", "CNN", "Deep Learning",
        "TensorFlow", "PyTorch", "Object Detection",
        "Image Processing"
    ],

    "Python Developer": [
        "Python", "OOP", "Django", "Flask",
        "FastAPI", "REST API", "SQL",
        "Git", "Testing", "Backend Development"
    ],

    "Backend Developer": [
        "Python", "Java", "Node.js", "FastAPI",
        "Django", "Express", "REST API",
        "PostgreSQL", "MongoDB", "Docker",
        "Authentication"
    ],

    "Full Stack Developer": [
        "HTML", "CSS", "JavaScript", "React",
        "Node.js", "Express", "MongoDB",
        "SQL", "REST API", "Git"
    ],

    "Software Engineer": [
        "Java", "Python", "C++",
        "Data Structures", "Algorithms", "OOP",
        "Git", "SQL", "System Design",
        "Software Engineering", "Testing"
    ],

    "DevOps Engineer": [
        "Docker", "Kubernetes", "Jenkins",
        "GitHub Actions", "Linux", "AWS",
        "CI CD", "Terraform", "Automation",
        "Monitoring"
    ],

    "Cloud Engineer": [
        "AWS", "Azure", "Google Cloud",
        "Docker", "Kubernetes", "Linux",
        "Networking", "Terraform",
        "Cloud Computing", "Cloud Security"
    ],

    "MLOps Engineer": [
        "Python", "Machine Learning", "Docker",
        "Kubernetes", "MLflow", "CI CD",
        "AWS", "Model Deployment",
        "Model Monitoring", "FastAPI"
    ],

    "Cybersecurity Analyst": [
        "Cybersecurity", "Network Security",
        "Linux", "Python", "Ethical Hacking",
        "Penetration Testing", "Firewalls",
        "Vulnerability Assessment", "SIEM"
    ],

    "Database Developer": [
        "SQL", "MySQL", "PostgreSQL",
        "Oracle", "MongoDB", "Database Design",
        "Query Optimization", "Stored Procedures",
        "Data Modeling"
    ]
}

training_rows = []

# Generate 80 varied examples per career
for role, skills in career_skill_map.items():

    for _ in range(80):

        # Randomly select 60%-100% of role skills
        min_skills = max(4, int(len(skills) * 0.6))

        selected = random.sample(
            skills,
            random.randint(min_skills, len(skills))
        )

        random.shuffle(selected)

        training_rows.append({
            "text": " ".join(selected),
            "role": role
        })

career_training_df = pd.DataFrame(training_rows)

# Shuffle dataset
career_training_df = career_training_df.sample(
    frac=1,
    random_state=42
).reset_index(drop=True)

print("Dataset shape:", career_training_df.shape)
print()
print(career_training_df["role"].value_counts())

career_training_df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    career_training_df["text"],
    career_training_df["role"],
    test_size=0.20,
    random_state=42,
    stratify=career_training_df["role"]
)

print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))
```

```python
model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)
```

```python
X_train_embeddings = model.encode(
    X_train.tolist(),
    normalize_embeddings=True,
    show_progress_bar=True
)

X_test_embeddings = model.encode(
    X_test.tolist(),
    normalize_embeddings=True,
    show_progress_bar=True
)

print(
    "Train embedding shape:",
    X_train_embeddings.shape
)

print(
    "Test embedding shape:",
    X_test_embeddings.shape
)
```

```python
from sklearn.linear_model import LogisticRegression

career_classifier = LogisticRegression(
    max_iter=2000,
    random_state=42
)

career_classifier.fit(
    X_train_embeddings,
    y_train
)

print(
    "Career classification model trained successfully!"
)
```

```python
from sklearn.metrics import (
    accuracy_score,
    classification_report
)

y_pred = career_classifier.predict(
    X_test_embeddings
)

accuracy = accuracy_score(
    y_test,
    y_pred
)

print(
    f"Model Accuracy: {accuracy * 100:.2f}%"
)

print("\nClassification Report:\n")

print(
    classification_report(
        y_test,
        y_pred,
        zero_division=0
    )
)
```

```python
def predict_career_ml(
    skills_text,
    top_n=5
):

    embedding = model.encode(
        [skills_text],
        normalize_embeddings=True
    )

    probabilities = (
        career_classifier.predict_proba(
            embedding
        )[0]
    )

    classes = (
        career_classifier.classes_
    )

    results = pd.DataFrame({
        "Career Role": classes,
        "Match Score": probabilities * 100
    })

    results = results.sort_values(
        "Match Score",
        ascending=False
    ).head(top_n)

    return results.reset_index(
        drop=True
    )
```

```python
def analyze_skill_gap(
    candidate_skills,
    target_role
):

    if target_role not in career_skill_map:

        return {
            "matched_skills": [],
            "missing_skills": [],
            "skill_match_percentage": 0
        }

    required_skills = career_skill_map[
        target_role
    ]

    candidate_text = (
        candidate_skills.lower()
    )

    matched = []

    missing = []

    for skill in required_skills:

        if skill.lower() in candidate_text:

            matched.append(skill)

        else:

            missing.append(skill)

    match_percentage = (
        len(matched)
        / len(required_skills)
    ) * 100

    return {

        "matched_skills":
            matched,

        "missing_skills":
            missing,

        "skill_match_percentage":
            round(match_percentage, 2)
    }
```

```python
candidate = """

Python
Machine Learning
Scikit-learn
Pandas
NumPy
FastAPI
Artificial Intelligence

"""

gap = analyze_skill_gap(
    candidate,
    "AI ML Engineer"
)

print(
    "Skill Match:",
    gap["skill_match_percentage"],
    "%"
)

print(
    "\nMatched Skills:"
)

print(
    gap["matched_skills"]
)

print(
    "\nMissing Skills:"
)

print(
    gap["missing_skills"]
)
```

```python
def careerlens_predict(
    candidate_skills,
    top_n=3
):

    # Generate embedding
    embedding = model.encode(
        [candidate_skills],
        normalize_embeddings=True
    )

    # Get probabilities
    probabilities = (
        career_classifier
        .predict_proba(
            embedding
        )[0]
    )

    classes = (
        career_classifier.classes_
    )

    results = pd.DataFrame({

        "Career Role":
            classes,

        "Match Score":
            probabilities * 100
    })

    results = results.sort_values(

        "Match Score",

        ascending=False

    ).head(top_n)

    results = results.reset_index(
        drop=True
    )

    # Best career
    best_role = (
        results.iloc[0][
            "Career Role"
        ]
    )

    confidence = float(
        results.iloc[0][
            "Match Score"
        ]
    )

    # Skill gap
    skill_gap = analyze_skill_gap(
        candidate_skills,
        best_role
    )

    return {

        "best_career":
            best_role,

        "confidence":
            round(confidence, 2),

        "top_careers":
            results,

        "matched_skills":
            skill_gap[
                "matched_skills"
            ],

        "missing_skills":
            skill_gap[
                "missing_skills"
            ],

        "skill_match":
            skill_gap[
                "skill_match_percentage"
            ]
    }
```

```python
candidate = """

Python
Machine Learning
Scikit-learn
Pandas
NumPy
FastAPI
Generative AI
Gemini
Ollama
React

"""

result = careerlens_predict(
    candidate
)

print(
    "BEST CAREER:",
    result["best_career"]
)

print(
    "MODEL CONFIDENCE:",
    result["confidence"],
    "%"
)

print(
    "SKILL COVERAGE:",
    result["skill_match"],
    "%"
)

print(
    "\nTOP CAREER RECOMMENDATIONS:"
)

display(
    result["top_careers"]
)

print(
    "\nMATCHED SKILLS:"
)

print(
    result["matched_skills"]
)

print(
    "\nSKILLS TO LEARN:"
)

print(
    result["missing_skills"]
)
```
